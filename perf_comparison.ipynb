```python
import subprocess
import matplotlib.pyplot as plt
import numpy as np
```

```python
default = []
packed = []
sizes = [i for i in range(1, 1001, 3)]
```

```python
for s in sizes:
    result = subprocess.check_output(['yarn', 'start', '--size', str(s)])
    default.append(float(result.decode('utf-8').split('\n')[2].split(': ')[-1]))
    packed.append(float(result.decode('utf-8').split('\n')[3].split(': ')[-1]))
```

```python
plt.plot(sizes, default, '-')
plt.plot(sizes, packed, '-')
plt.show()
```

```python
plt.plot(sizes, np.array(packed)/np.array(default), '-')
plt.show()
```
